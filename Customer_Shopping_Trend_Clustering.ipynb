```python
# Function to create heatmap for a given categorical column
def create_heatmap(df, category_column, value_column):
    # Aggregate the data
    agg_data = df.groupby([category_column])[value_column].mean().reset_index()

    # Create heatmap
    fig = px.imshow(agg_data.pivot_table(index=[category_column], values=value_column),
                    labels=dict(color=value_column),
                    title=f'{value_column} by {category_column}',
                    aspect="auto")
    fig.show()

# List of categorical columns to create heatmaps for
categorical_columns = ['Gender', 'Item Purchased', 'Category', 'Size', 'Color', 'Season']

# Create heatmaps for each categorical column
for column in categorical_columns:
    create_heatmap(df, column, 'Review Rating')
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
def cramers_v(x, y):
    """Calculate Cramer's V statistic for categorical-categorical association."""
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Calculate Cramer's V for each pair of categorical columns
columns = ['Gender', 'Item Purchased', 'Category', 'Size', 'Color', 'Season']
cramers_v_matrix = pd.DataFrame(index=columns, columns=columns)

for col1 in columns:
    for col2 in columns:
        cramers_v_matrix.loc[col1, col2] = cramers_v(df[col1], df[col2])

# Convert values to numeric
cramers_v_matrix = cramers_v_matrix.astype(float)

# Display the correlation matrix
print(cramers_v_matrix)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Drop 'Customer ID' if it exists in your data
df_features = df.drop(columns=['Customer ID'])

# Standardize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_features)
```

```python
# Define a range of cluster numbers
k_range = range(1, 11)  # You can adjust the range if needed
inertia = []

# Fit K-Means and calculate inertia for each k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve for K-Means Clustering')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

# Define a range of cluster numbers
k_range = range(2, 11)  # Starting from 2 because silhouette score is not defined for 1 cluster
silhouette_scores = []

# Fit K-Means and calculate silhouette score for each k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    labels = kmeans.labels_
    score = silhouette_score(df_scaled, labels)
    silhouette_scores.append(score)

# Plot the silhouette scores
plt.figure(figsize=(8, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.grid(True)
plt.show()
```

```python
# Choose the optimal number of clusters (e.g., 4 from the elbow curve)
optimal_k = 2

# Fit K-Means with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# Add cluster labels to the original DataFrame
df['Cluster'] = clusters

```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import numpy as np

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Try different eps and min_samples
eps_values = np.arange(0.1, 5.0, 0.1)
min_samples_values = range(2, 20)

best_silhouette_score = -1
best_eps = None
best_min_samples = None

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_scaled)

        if len(set(labels)) > 1:  # Ensure there are clusters (not all noise)
            mask = labels != -1
            X_clustered = X_scaled[mask]
            labels_clustered = labels[mask]

            try:
                silhouette_avg = silhouette_score(X_clustered, labels_clustered)
                if silhouette_avg > best_silhouette_score:
                    best_silhouette_score = silhouette_avg
                    best_eps = eps
                    best_min_samples = min_samples
            except:
                continue

print(f"Best Silhouette Score: {best_silhouette_score:.3f}")
print(f"Best eps: {best_eps}")
print(f"Best min_samples: {best_min_samples}")
```
